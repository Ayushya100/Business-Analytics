## 3. Data Cleaning
### 3.1 Handling Outliers

```python
store.columns
```

#### Sales Column

```python
store.Sales.describe()
```

```python
store.Sales.quantile([.25, .50, .75, .90, .95, .99])
```

```python
plt.figure(figsize=(18,4))
sns.boxplot(x = store['Sales'])
plt.show()
```

We found that their are some outliers present in Sales attribute and we need to handle it for our analysis not to get affected

We will delete all those rows whose Sales values which are beyond 5000

```python
store = store[store['Sales'] < 2000]
```

```python
store.Sales.describe()
```

#### Quantity Column

```python
store.Quantity.describe()
```

```python
store.Quantity.quantile([.25, .50, .75, .90, .95, .99])
```

```python
plt.figure(figsize=(18,4))
sns.boxplot(x = store['Quantity'])
plt.show()
```

There are some Outlier present in our Quantity attribute but it will not affect much in our analysis, even we can simply delete these outliers from our datasets because they are less in quantity

```python
store = store[store['Quantity'] < 11]
```

```python
store.Quantity.describe()
```

#### Discount Column

```python
store.Discount.describe()
```

```python
store.Discount.quantile([.25, .50, .75, .90, .95, .99])
```

```python
plt.figure(figsize=(18,4))
sns.boxplot(x = store['Discount'])
plt.show()
```

These data points that seems like an outlier are actually lies in the 99 percentile so we are not required to remove them or to impute values for our analysis

#### Profit Column

```python
store.Profit.describe()
```

```python
store.Profit.quantile([.25, .50, .75, .90, .95, .99])
```

```python
plt.figure(figsize=(18,4))
sns.boxplot(x = store['Profit'])
plt.show()
```

We have outliers present in Profit column but we will not remove them as they are the columns for which we are required to find why the overall profit is so low.

#### Sales Vs Discount

```python
plt.figure(figsize=(18, 6))
sns.scatterplot(x = store['Sales'], y = store['Discount'], color = 'red')
plt.show()
```

```python
sns.pairplot(data = store, vars=['Sales', 'Quantity', 'Discount', 'Profit'])
plt.show()
```

```python
store[['Sales', 'Quantity', 'Discount', 'Profit']].corr()
```

```python
sns.heatmap(store[['Sales', 'Quantity', 'Discount', 'Profit']].corr(), annot = True, cmap='Reds')
```

There is a high Negative Correlation between Profit and Discount which means that, when we increase the discount (more than 30% especially) it cause us more loss than profit.

Sales also gets lower when we increase Discount (more than 40% especially) on our products.

### Numerical - Categorical Analysis

```python
store.columns
```

#### Ship Mode Vs Sales

```python
store.groupby('Ship Mode')['Sales'].mean()
```

```python
store.groupby('Ship Mode')['Sales'].median()
```

```python
plt.figure(figsize=(18,6))
sns.boxplot(x = 'Sales', y = 'Ship Mode', data = store)
plt.show()
```

#### Ship Mode Vs Profit

```python
store.groupby('Ship Mode')['Profit'].mean()
```

```python
store.groupby('Ship Mode')['Profit'].median()
```

```python
store.groupby('Ship Mode')['Profit'].aggregate(['mean', 'median']).plot(kind = 'barh')
plt.show()
```

Even though we have a very low Sale on Shipping mode as 'Same Day' but still we have a very high Profit. So, we should increase the Sales on the Same Day.

```python
store.groupby('Segment')['Profit'].median()
```
